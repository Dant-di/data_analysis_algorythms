1. Подберите скорость обучения (eta) и количество итераций

```python
import numpy as np
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
```

```python
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
def calc_mae(y, y_pred):
    err = np.mean(np.abs(y - y_pred))
    return err

def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
W_analytical = np.linalg.inv(np.dot(X.T, X)) @ X.T @ y
W_analytical
```

```python
# оригинальный код
n = X.shape[0]

eta = 1e-2
n_iter = 100

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

```python
# модификации
n = X.shape[0]
# Подбор параметров сделан в несколько этапов:

# 1. Количество итераций увеличено до 1000. Изменения весов замедлились при 700 итерациях
# 2. eta установленно в 0.05. Изменения весов замедлились при 380
# 3. Частота корреции скорости обчения была снижена до каждой 50 итерации, значение корректировки снижено до 1.01. Целевые значения, полученные аналитическим путем, были достигнуты на 550 итерации.
# 5. Отключена корректировка скорости обчения, количество итераций выставлено в 500, частота вывода результат установлена назад на 10
# 6. Целевые значения были достигнуты на 490 итерации.

eta = 0.05


n_iter = 500

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))

    if i % 10 == 0:
        # eta /= 1.01
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

```python
eta = 1e-2

W = np.array([1, 0.5])
W_new = np.copy(W)
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')
limit = 0.001

while True:
    y_pred = np.dot(X, W_new)
    err = calc_mse(y, y_pred)
    for k in range(W_new.shape[0]):
        W_new[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    dif = W - W_new
    if abs(dif[0]) > limit and abs(dif[1]) > limit:
        W = np.copy(W_new)
        print(f'W_new = {W_new}, MSE = {round(err, 2)}')
        continue
    else:
        print('Script ended')
        break

```
